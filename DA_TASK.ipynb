```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_excel("/content/DA_Task1.xlsx")
print(df)
```

1. Data Cleaning:

```python
# check for the missing values

m_v=df.isnull().sum()

# check the duplicated rows

d_r=df.duplicated().sum()

print(m_v,d_r)
```

‚úÖ Step 1: Data Cleaning Completed
üìå Missing Values: 0
üìå Duplicate Rows: 0

 2. Exploratory Data Analysis (EDA):

```python
# summary statistics of sales and profit
s_s=df[["Sales","Profit"]].describe()
print(s_s)
```

```python
# distrbution of sales
plt.figure(figsize=(10,5))
sns.histplot(df["Sales"],bins=30,kde=True,color="blue")
plt.title("Distribution of Sales")
plt.xlabel("Sales Amount")
plt.ylabel("Frequency")
plt.show()
```

‚úÖ Step 2: Exploratory Data Analysis (EDA) Completed

3. Sales Performance Analysis:

```python
# total sales, quantity,and profit
t_s=df['Sales'].sum()
t_q=df['Quantity'].sum()
t_p=df['Profit'].sum()

print(t_s,t_q,t_p)
```

```python
# Average sales and profit per order
a_s=df['Sales'].mean()
a_p=df['Profit'].mean()

print(a_s,a_p)
```

4. visulization

```python
# visulization - sales and profit
plt.figure(figsize=(10,5))
sns.scatterplot(x=df["Sales"],y=df["Profit"],alpha=0.5,color="red")
plt.title("Sales vs Profit")
plt.xlabel("Sales Amount")
plt.ylabel("Profit Amount")
plt.show()
```
